```python
import numpy as np
import matplotlib.pyplot as plt

# Definici칩n de puntos de dominio

N_points_domain = 100

# Definici칩n de rango

x_range = np.linspace(-2.5,2.5, N_points_domain) # rango se pone -2.5 y 2.5

# Definici칩n de funciones

def funcion_y(x):
    result = (x**2-4)**3 + 2
    return result

def funcion_derivada(x):
    result = 3*(x**2-4)**2 * 2*x # ojo syntax 2x es 2*x
    return result 

# Gr치ficos

plt.plot(x_range, funcion_y(x_range))
plt.plot(x_range, funcion_derivada(x_range))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definici칩n de puntos de dominio

N_points_domain = 100

# Definici칩n de rango

x_range = np.linspace(-2.5,2.5, N_points_domain) # rango se pone -2.5 y 2.5

def funcion_y(x):
    result = (x**2-4)**3 + 2
    return result

# con la regla de la cadena adentro
# aqu칤 se hizo el forward con cada uno de los pasos de la regla de la cadena
def funcion_reglacadena(x):
    t=x**2
    u=t-4
    z=u**3
    y=z+2
    result = 1*3*u**2*1*2*x # ojo syntax 2x es 2*x
    return result 

# Gr치ficos

plt.plot(x_range, funcion_y(x_range))
plt.plot(x_range, funcion_reglacadena(x_range))

# Sobre funciones elementales se obtiene lo mismo.
```
